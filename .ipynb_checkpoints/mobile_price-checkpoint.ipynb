```python
def missing_data(df):
    total_missing = df.isnull().sum().sort_values(ascending = False)
    Percentage = (total_missing / len(df)) * 100
    summary_df = pd.concat([total_missing, Percentage], axis=1, keys=['Data NUM Missing', 'Data Percentage Missing'])
    return summary_df

missing_data_df = missing_data(df)
missing_data_df.style.background_gradient(cmap='Greys')
```

```python
df["Price in INR"] = df["Price in INR"].str.replace("₹", "").str.replace(",", "") 
df["Price in INR"] = df["Price in INR"].astype("int") 
df["Number of Ratings"] = df["Number of Ratings"].str.replace(",", "")
df["Number of Ratings"] = df["Number of Ratings"].astype("int")
```

```python
df.rename(columns={'Rating ?/5':'Rating 0/5'},inplace=True)
```

```python
df["Battery"] = df["Battery"].str.replace("mAh", "")
lst=['A15 ', 'A13 ', 'A14 ', 'A16 ', 'A12 ', 'A9 ','Apple ', '1 ','0 ','MediaTek ', 'Brand ','Unisoc ','2 ']
a=0
for j in range(0,len(lst)):
    for i in range(0,len(df['Battery'].values)):
        if (df['Battery'][i]== lst[j]):
            df['Battery'][i]=4000
df['Battery'].fillna(4000,inplace=True)
df["Battery"] = df["Battery"].astype('int')
```

```python
df['Front Camera'].fillna('No Value Available',inplace=True)
df['ROM/Storage'].fillna('No Value Available',inplace=True)
df['Processor'].fillna('No Value Available',inplace=True)
df['Back/Rare Camera'].fillna('No Value Available',inplace=True)
```

```python
df.isnull().sum()

```

```python
df['Phone Name'] = df['Phone Name'].astype(str)
df['BRAND'] = df['Phone Name'].str.split().str[0]
```

```python
df['ROM/Storage'] = df['ROM/Storage'].astype(str)
for i in range(0,1836):
    if df['ROM/Storage'][i]== 'No Value Available':
        df['ROM(GB)'] = 0
    else:
        df['ROM(GB)'] = df['ROM/Storage'].str.split().str[0]
rom_list=['No', '0', 'Expandable', 'NA']
for i in range(0,1836):
    if df['ROM(GB)'][i] in rom_list:
        df['ROM(GB)'][i]='0'
```

```python
df['RAM'] = df['RAM'].astype(str)
for i in range(0,1836):
    if df['RAM'][i]== 'No Value Available':
        df['RAM(GB)'] = 0
    else:
        df['RAM(GB)'] = df['RAM'].str.split().str[0] 
ram_list=['NA', 'Expandable','0.53','3.81', 'cm','0.046875','1.5']
for i in range(0,1836):
    if df['RAM(GB)'][i] in ram_list:
        if df['RAM(GB)'][i] == '0.53':
            df['RAM(GB)'][i]='1'
        elif df['RAM(GB)'][i] == '3.81':
            df['RAM(GB)'][i]='4'
        elif df['RAM(GB)'][i] == '0.046875':
            df['RAM(GB)'][i]='0'
        elif df['RAM(GB)'][i] == '1.5':
            df['RAM(GB)'][i]='2'
        else:
            df['RAM(GB)'][i]='0'
```

```python
df['ROM(GB)']=df['ROM(GB)'].astype('float')
df['RAM(GB)']=df['RAM(GB)'].astype('int')
```

```python
BackCamera_list=[]
for i in range(0,len(df['Back/Rare Camera'])):
    s=(df['Back/Rare Camera'][i].split("+"))
    BackCamera_list.append(s)
score=[]
for i in range(0,len(BackCamera_list)):
    sr=len(BackCamera_list[i])
    score.append(sr)
df['Back_camera_Score']=score
```

```python
frontCamera_list=[]
for i in range(0,len(df['Back/Rare Camera'])):
    s=(df['Front Camera'][i].split("+"))
    frontCamera_list.append(s)
score1=[]
for i in range(0,len(frontCamera_list)):
    sr=len(frontCamera_list[i])
    score1.append(sr)
df['front_camera_Score']=score1
```

```python
df['Main_Back_Camera'] = df['Back/Rare Camera'].str.split().str[0] 
df['Main_Back_Camera'] = df['Main_Back_Camera'].str.replace("MP", "")
df['Main_Back_Camera'] = df['Main_Back_Camera'].str.replace("Mp","")
```

```python
df['Main_Front_Camera'] = df['Front Camera'].str.split().str[0] 
df['Main_Front_Camera'] = df['Main_Front_Camera'].str.replace("MP", "").replace("Mp","")
```

```python
df['Price_Range']=0
a=0
for i in range(0,len(df['Price in INR'].values)):
    if df['Price in INR'][a] <= 7000:
        df['Price_Range'][a]=1
    elif df['Price in INR'][a]  <= 14000:
        df['Price_Range'][a]=2
    elif df['Price in INR'][a]  <= 21000:
        df['Price_Range'][a]=3
    elif df['Price in INR'][a]  <= 30000:
        df['Price_Range'][a]=4
    elif df['Price in INR'][a] <= 45000:
        df['Price_Range'][a]=5
    elif df['Price in INR'][a] <= 60000:
        df['Price_Range'][a]=6
    else:
        df['Price_Range'][a]=7
    a=a+1
```

```python

#o Feature Selection  
features = ["Rating 0/5", "Number of Ratings", "RAM(GB)", "ROM(GB)", "Battery", "Encoding_Processor","Price in INR",
            "Back_camera_Score", "front_camera_Score", "Main_Back_Camera", "Main_Front_Camera", "Price_Range","Encoding_Brand"]

# Selecting features and target variable
X = df[features]
y = df["Price in INR"]

# Splitting the dataset into training and testing sets( Data Split )
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

 

# Initializing the Decision Tree regressor
regressor = DecisionTreeRegressor()

# Fitting the regressor on the training data
regressor.fit(X_train, y_train)

# Predicting the prices for test set
y_pred = regressor.predict(X_test)

# Convert the prediction to integers
y_pred_int = y_pred.astype(int)

# Calculate precision, recall, and accuracy
precision = precision_score(y_test, y_pred_int, average='weighted')
recall = recall_score(y_test, y_pred_int, average='weighted')
accuracy = accuracy_score(y_test, y_pred_int)

print("Precision:", precision)
print("Recall:", recall)
print("Accuracy:", accuracy)

import matplotlib.pyplot as plt

# Plotting the predicted vs. actual prices
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_int, color='blue', label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', lw=2, label='Ideal Predictions')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Decision Tree Regressor: Actual vs. Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()
 # Plotting the residuals
residuals = y_test - y_pred_int

plt.figure(figsize=(10, 6))
plt.scatter(y_test, residuals, color='green')
plt.xlabel('Actual Price')
plt.ylabel('Residuals')
plt.title('Decision Tree Regressor: Residual Plot')
plt.axhline(y=0, color='red', linestyle='--', lw=2)
plt.grid(True)
plt.show()

```

```python


# Assuming df is your DataFrame
# Split the data into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.25, random_state=100)

print("Number of rows in training set:", len(train_data))
print("Number of rows in testing set:", len(test_data))

# Prepare the data for modeling
train_data_model = df[['BRAND', 'Encoding_Brand', 'Number of Ratings', 'Battery', 'Rating 0/5', 'ROM(GB)', 'RAM(GB)',
                       'Back_camera_Score', 'front_camera_Score', 'Main_Back_Camera', 'Main_Front_Camera','Price in INR']]
train_data_model.drop(columns='BRAND', axis=1, inplace=True)
train_data_model_X = train_data_model.iloc[:, :-1]
train_data_model_Y = train_data_model.iloc[:, -1]

# Standardize the features
columns = ['Encoding_Brand', 'Battery', 'Number of Ratings', 'Rating 0/5', 'ROM(GB)', 'RAM(GB)', 'Back_camera_Score'
           ,'front_camera_Score','Main_Back_Camera', 'Main_Front_Camera']
CT = ColumnTransformer(transformers=[('SS', StandardScaler(), columns)], remainder='passthrough')
train_data_model_X = CT.fit_transform(train_data_model_X)

# Split the standardized data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(train_data_model_X, train_data_model_Y, test_size=0.5, random_state=100)

# Train the Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)
y_pred_lr_normal = model.predict(x_test)

# Evaluate the model using R-squared score
print('R-squared score without using grid search CV:', r2_score(y_test, y_pred_lr_normal))

# Define a meaningful threshold for classification
threshold = np.mean(np.abs(y_test - np.mean(y_test)))  # Mean absolute deviation

# Classify predictions as True if within threshold, False otherwise
y_pred_class = np.abs(y_pred_lr_normal - y_test) <= threshold
y_test_class = np.ones_like(y_test)  # Assuming all true values are within threshold for simplicity

# Calculate precision, recall, and accuracy
precision = precision_score(y_test_class, y_pred_class)
recall = recall_score(y_test_class, y_pred_class)
accuracy = accuracy_score(y_test_class, y_pred_class)

print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'Accuracy: {accuracy}')

# Perform hyperparameter tuning using GridSearchCV
param_grid = {'fit_intercept': [True, False], 'positive': [True, False]}
grid_search_lr = GridSearchCV(model, param_grid, cv=5, n_jobs=-1)
grid_search_lr.fit(x_train, y_train)

# Evaluate the best model from GridSearchCV
best_model = grid_search_lr.best_estimator_
y_pred_best = best_model.predict(x_test)

print('R-squared score with GridSearchCV:', r2_score(y_test, y_pred_best))

# Convert the best model's predictions to classification
y_pred_best_class = np.abs(y_pred_best - y_test) <= threshold

# Calculate precision, recall, and accuracy for the best model
precision_best = precision_score(y_test_class, y_pred_best_class)
recall_best = recall_score(y_test_class, y_pred_best_class)
accuracy_best = accuracy_score(y_test_class, y_pred_best_class)

print(f'Best Model Precision: {precision_best}')
print(f'Best Model Recall: {recall_best}')
print(f'Best Model Accuracy: {accuracy_best}')

import matplotlib.pyplot as plt
import numpy as np

# Plotting the predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_lr_normal, color='blue', label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', lw=2, label='Ideal Predictions')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Linear Regression: Actual vs. Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the residuals
residuals = y_test - y_pred_lr_normal
plt.figure(figsize=(10, 6))
plt.scatter(y_test, residuals, color='green')
plt.xlabel('Actual Price')
plt.ylabel('Residuals')
plt.title('Linear Regression: Residual Plot')
plt.axhline(y=0, color='red', linestyle='--', lw=2)
plt.grid(True)
plt.show()
```
